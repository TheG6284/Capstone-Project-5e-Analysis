```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #for plotting
import seaborn as sns

df = pd.DataFrame(pd.read_csv('MonsterData.csv'))
df.info()
```

```python
cr_bar = sns.barplot(x = df.value_counts('CR').index, y = df.value_counts('CR').values)

x = cr_bar.get_figure()

x.savefig("BarPlot_CR.png")

plt.clf()
```

```python
#plt.figure(figsize=(14,4)) # this creates a figure 8 inch wide, 4 inch high
cr_bar = sns.countplot(x="CR", data=df)
cr_bar.set_xticklabels(cr_bar.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()

x = cr_bar.get_figure()

x.savefig("BarPlot_CR.png")

plt.clf()
```

```python

size_bar = sns.countplot(x="Size", data=df)
plt.show()

x = size_bar.get_figure()
x.savefig("BarPlot_size.png")

plt.clf()

type_bar = sns.countplot(x="Type", data=df)
type_bar.set_xticklabels(type_bar.get_xticklabels(), rotation=40, ha="right")
plt.tight_layout()
plt.show()

x = type_bar.get_figure()

x.savefig("BarPlot_type.png")

plt.clf()
```

```python
scatter_ac = sns.scatterplot(df, x = "Size", y = "AC", hue = "CR")

x = scatter_ac.get_figure()
x.savefig("scatter_ac.png")

plt.clf()

scatter_hp = sns.scatterplot(df, x = "Size", y = "HP", hue = "CR")

x = scatter_hp.get_figure()
x.savefig("scatter_hp.png")

plt.clf()
```

```python
facet_size = sns.FacetGrid (df, col = "Size", hue = "CR", col_wrap = 2)
facet_size.map (sns.scatterplot, "CR", "Type")
facet_size.add_legend()
plt.show ()


facet_size.savefig("facet_size.png")


plt.clf()
```

```python


heatmap_stats = sns.heatmap(df.corr(), annot = True)



x = heatmap_stats.get_figure()    

x.savefig("heatmap_stats.png")


```
